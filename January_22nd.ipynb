```python
def production_and_mps(K, L, A = 1, alpha = 0.3, beta = 0.7):
  Y = A * (K ** alpha) * (L ** beta)
  MPK = A * alpha * (K ** (alpha - 1)) * (L ** beta)
  MPL = A * beta * (K ** alpha) * (L ** (beta - 1))
  return {
      "output": Y,
      "MPK": MPK,
      "MPL": MPL
  }
```

```python
results = production_and_mps(10, 20)
print(results)
print(f"prodcution: {results['output']}")
print(f"MPK: {results['MPK']}")
print(f"MPL: {results['MPL']}")
```

```python
def cobb_douglas(K, L):
    """
    Computes the production F(K, L) for a Cobb-Douglas production function

    Takes the form F(K, L) = A K^{\alpha} L^{1 - \alpha}

    We restrict A = 1 and alpha = 0.33
    """
    return 1.0 * K**(0.33) * L**(1.0 - 0.33)
```
