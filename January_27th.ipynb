```python
class BankAccount:
  def __init__(self, balance):
    self.balance = balance # attribute (data)

  def deposit(self, amount):
    self.balance += amount # method (function) uses the data

# Create an object (instance) of BankAccount
account_1 = BankAccount(100) # 'account' is the object, starting balance = 100
account_2 = BankAccount(200)

# Call the deposit method on the object to update balance
account_1.deposit(50)
account_2.deposit(30)

# Check the updated balance
print(account_1.balance) # Output: 150
print(account_2.balance) # Output: 230
```

A 3D array is a data structure that stores elements in three dimensions—depth (or layers), rows, and columns—essentially forming a cube of data. It can be visualized as a collection of multiple 2D arrays stacked together, similar to stacking multiple trays on top of each other.

An element in a 3D array arr is accessed using the notation arr[i][j][k], where i is the depth, j is the row, and k is the column.

A 3D array can be declared by specifying all three dimensions.

For example:
int arr[4][5][8] declares a 3D integer array with 4 layers, 5 rows per layer, and 8 columns per row
This array can store a total of 4 × 5 × 8 = 160 elements
