```python
def mean(numbers):
  return sum(numbers) / len(numbers)

print(mean([1, 2, 3]))
```

```python
class BankAccount:
  def __init__(self, balance):
    self.balance = balance # attribute (data)

  def deposit(self, amount):
    self.balance += amount # method (function) uses the data

# Create an object (instance) of BankAccount
account_1 = BankAccount(100) # 'account' is the object, starting balance = 100
account_2 = BankAccount(200)

# Call the deposit method on the object to update balance
account_1.deposit(50)
account_2.deposit(30)

# Check the updated balance
print(account_1.balance) # Output: 150
print(account_2.balance) # Output: 230
```
