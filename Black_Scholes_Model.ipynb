The Black-Scholes Model is a mathematical formula used to calculate the theoretical price of options contracts, specifically European-style options. Developed by economists Fischer Black, Myron Scholes, and Robert Merton, the model revolutionized options pricing by providing a systematic way to value derivatives based on several key variables.
