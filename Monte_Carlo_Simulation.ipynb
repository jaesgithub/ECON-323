# Monte Carlo Simulation?
Monte Carlo simulation uses repeated random sampling to estimate outcomes in uncertain situations. It runs a model thousands of times with different random inputs to predict a range of possible results and their probabilities.
### Core Idea
This method relies on the law of large numbers, where averaging many random trials approximates the true expected value. Computers generate random numbers from probability distributions for uncertain variables, like stock returns or physical measurements.
### How It Works
Define a model with inputs that have uncertainty, assign probability distributions to them, and simulate thousands of scenarios by randomly sampling those inputs each time. Calculate the output for each run, then analyze the distribution of results to understand averages, risks, and extremes.
### Example
To estimate Ï€, generate random points in a unit square and count the proportion inside a quarter circle; multiply by 4 for the approximation, which improves with more points. This shows how random sampling estimates hard-to-compute values like areas or integrals.
### Common Uses
Finance applies it for portfolio risk (e.g., Value at Risk), engineering for reliability testing, and physics for particle simulations. It handles complex problems where exact math fails.

```python
import numpy as np
simulations = 1000000

A = np.random.uniform(1, 5, simulations)
B = np.random.uniform(2, 6, simulations)

duration = A + B
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (3, 1.5))
plt.hist(duration, density = True)
plt.axvline(9, color = 'r')
plt.show()
print((duration > 9).sum()/simulations)
```
