```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader as web
import datetime as dt
```

```python
fig, ax = plt.subplots()

x = np.linspace(0, 2*np.pi, 100) # 100 evenly spaced points between 0 and 2 pi
y = np.sin(x)

ax.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 3, 100)
fig, ax = plt.subplots(figsize = (6, 4), layout = 'constrained')
ax.plot(x, x, label = 'Linear')
ax.plot(x, x**2, label = 'Quadratic')
ax.set_xlabel('X value')
ax.set_ylabel('Y value')
ax.set_title('Simple Plot')
ax.legend
plt.show()
```

```python
np.random.seed(123)
data = {'a': np.arange(50),    # 50 integers from 0 to 49
        'c': np.random.randint(0, 50, 50),  # 50 random integers from 0 to 49
        'd': np.random.randn(50)}     # 50 normally distributed random values
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize = (6,4), layout = 'constrained')
ax.scatter('a', 'b', c = 'c', s ='d', data = data)
# c = color of points and s = size of points
ax.set_xlabel('value of a')
ax.set_ylabel('value of b')
plt.show()
```

```python
data1, data2 = np.random.randn(2, 100)  # two sets of data
fig, ax = plt.subplots(figsize = (6, 3))
x = np.arange(len(data1))
ax.plot(x, np.cumsum(data1), c = "blue", lw = 3, ls ='--')
ax.plot(x, np.cumsum(data2), c = 'orange', lw = 2)
plt.show()
```
