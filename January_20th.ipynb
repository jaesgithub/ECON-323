```python
def mean(numbers):
  total = sum(numbers)
  N = len(numbers)
  answer = total / N

  return answer
```

```python
x = [1, 2, 3, 4]
the_mean = mean(x)
print(the_mean)
```

```python
def cobb_douglas(K, L):

  # Create alpha and z
  A = 1
  alpha = 0.33

  return A * K**alpha * L**(1 - alpha)
```

```python
def cobb_douglas(K, L):

  # Create alpha and z
  A = 1
  alpha = 0.33
  beta = 2

  return A * K**alpha * L**beta
```

```python
cobb_douglas(2.0, 2.0)
```

```python
def returns_to_scale(K, L):
  y1 = cobb_douglas(K, L)
  y2 = cobb_douglas(2*K, 2*L)
  y_ratio = y2/y1
  return y_ratio
```

```python
returns_to_scale(4.0, 5.0)
```
