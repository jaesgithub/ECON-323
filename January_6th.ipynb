```python
tu = (1, 2, 3)
```

```python
type(tu)
```

### Accessing elements
You can access individual elements of a tuple using indexing (starting from 0) or a range of elements using slicing.

```python
print(f"Original tuple: {tu}")

# Accessing an element by index
print(f"First element: {tu[0]}")

# Slicing the tuple
print(f"Elements from index 1 to the end: {tu[1:]}")
```

### Concatenation
You can combine two or more tuples using the `+` operator.

```python
tuple_new = tu + (4, 5, 6)
print(f"Concatenated tuple: {tuple_new}")
```

### Repetition
You can repeat a tuple using the `*` operator.

```python
repeated_tuple = tu * 2
print(f"Repeated tuple: {repeated_tuple}")
```

### Membership testing
You can check if an element exists in a tuple using the `in` operator.

```python
print(f"Is 2 in tu? {2 in tu}")
print(f"Is 5 in tu? {5 in tu}")
```

### Tuple length
You can find the number of elements in a tuple using the `len()` function.

```python
print(f"Length of tu: {len(tu)}")
```
