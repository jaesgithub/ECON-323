Prints only if statement is True.

```python
val = True

if val is True: # Check an expression
  print("code is run")
  print("More codes run") # Inside if block

print("Code runs after 'if' block, regardless of val") # Outside if block
```

```python
val = True

if val is False: # Check an expression
  print("code is run")
  print("More codes run") # Inside if block

print("Code runs after 'if' block, regardless of val") # Outside if block
```

```python
if (1 < 2) is True:
  print("1 < 2")

if 1 < 2:
  print("1 < 2")
```

```python
import datetime # datetime(year, month, date)
halloween = datetime.date(2017, 10, 31)

if halloween.month > 9:
    print("Halloween is in Q4")
elif halloween.month > 6:
    print("Halloween is in Q3")
elif halloween.month > 3:
    print("Halloween is in Q2")
else:
    print("Halloween is in Q1")
```
