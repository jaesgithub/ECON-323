```python
company_revenue = [5.12, 5.20, 5.50, 6.50]

for index, value in enumerate(company_revenue):
    print(f"Month {index + 1} revenue was: {value} billion dollars")
```

```python
company_revenue = [5.12, 5.20, 5.50, 6.50]

for index, value in enumerate(company_revenue):
    print(f"quarter {index} revenue is: ${value} million")
```

The enumerate() function in Python adds a counter to an iterable and returns an enumerate object, which is an iterator of tuples containing index-value pairs. It simplifies looping over sequences when both the position and value are needed.

The function uses enumerate(iterable, start=0), where iterable is any sequence like a list or tuple, and start sets the initial index (defaults to 0).

```python
enumerate(company_revenue)
```

```python
list(enumerate(company_revenue))
```

```python
x = list(range(4))

# Create squared values with a loop
x2_loop = []
for x_val in x:
    x2_loop.append(x_val**2)

# Create squared values with a comprehension
x2_comp = [x_val**2 for x_val in x]

print(x2_loop)
print(x2_comp)
```

```python
# Create a dictionary from lists
tickers = ["AAPL", "GOOGL", "TVIX"]
prices = [175.96, 1047.43, 8.38]
d = {key: value for key, value in zip(tickers, prices)}
d
```

```python
# Create a list from a dictionary
d = {"AMZN": "Seattle", "TVIX": "Zurich", "AAPL": "Cupertino"}

hq_cities = [d[ticker] for ticker in d.keys()]
hq_cities
```
