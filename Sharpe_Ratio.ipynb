# Definition
The Sharpe Ratio measures how much extra return an investment provides for each unit of risk taken, helping investors compare performance beyond just raw gains. Developed by William Sharpe, it focuses on risk-adjusted returns rather than total returns alone.

# Interpretation
A higher ratio (e.g., above 1) indicates better risk-adjusted performance; 1-2 is good, 2-3 very good, over 3 excellent.
Negative values mean returns don't beat the risk-free rate after accounting for volatility.
It assumes risk is fully captured by standard deviation, so it's best for diversified portfolios.

```python
def sharpe_ratio(return_series, N, rf):
    mean = return_series.mean() * N - rf
    sigma = return_series.std() * np.sqrt(N)
    return mean / sigma
```

```python
import numpy as np
import pandas as pd
import yfinance as yf

# Download stock data
stock_data = yf.download('AAPL', '2020-01-01', '2024-01-01')

# Calculate logarithmic returns
returns = np.log(stock_data['Close'] / stock_data['Close'].shift(1))

# Calculate Sharpe Ratio
volatility = returns.std() * np.sqrt(252)
sharpe_ratio = ((returns.mean() * 252) - 0.05) / volatility
print(sharpe_ratio)
```
